```python
import pydicom as dicom                         #code to open DICOM files in python. 
import matplotlib.pylab as plt

imgp = 'MR0002.dcm'
ds = dicom.dcmread(imgp)

plt.imshow(ds.pixel_array)
```

```python
import matplotlib.pyplot as plt     # Greyscaling the DICOM image 
import pydicom                      # Greyscaling images can be the result of measuring the intensity of light at each
                                    # pixel according to a particular weighted combination of frequencies. 
imgp = 'MR0002.dcm'
ds = pydicom.dcmread(imgp)
plt.imshow(ds.pixel_array, cmap=plt.cm.bone)     
ds.PixelData
plt.gray()
plt.show()

print()
print()
```
